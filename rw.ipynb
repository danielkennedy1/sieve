```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('market_history.json', 'r') as f:
    data = json.load(f)

data["Generations"] = 0
```

```python
df = pd.DataFrame.from_dict(data)
```

```python
df.head()
```

```python
df[["Prices", "Volumes", "Timestamps"]].iloc[:100].plot(
    x='Timestamps',
    subplots=True,
    title='Initial Generation Market Behaviour',
    )
```
